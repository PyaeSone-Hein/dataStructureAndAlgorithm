```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

        return None

    def is_empty(self):
        return len(self.items) == 0

def is_balanced(input_string):
    stack = Stack()
    bracket_map = {')' : '(', ']' : '[', '}' : '{'}

    for char in input_string:
        if char in '({[':
            stack.push(char)

        elif char in ')}]':
            top = stack.pop()
            if top is None or bracket_map[char] != top:
                return False

    return stack.is_empty()

print(is_balanced("()"))      
print(is_balanced("([{}])")) 
print(is_balanced("([)]"))
print(is_balanced("([{})"))
```
