```python
class Node:
    def __init__(self, data = None):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.is_empty():
            return 'Stack is empty'

        else:
            popped_data = self.top.data
            self.top =  self.top.next
            return popped_data

s = Stack()
print(s.is_empty())

s.push(10)
s.push(20)
print(s.is_empty())

print(s.pop())
print(s.pop())
print(s.pop())
```
