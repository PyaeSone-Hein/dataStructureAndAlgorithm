```python
myArray = np.array([10, 20, 30,40, 50, 8, 6])
if myArray.size == 0:
    print('The array is empty')
else:
    min_val = myArray[0]
    max_val = myArray[0]
    for i in myArray[1:]:
        if i < min_val:
            min_val = i
        elif i > max_val:
            max_val = i
    print(f'Minimun: {min_val}')
    print(f'Maximum: {max_val}')
```

```python
#move zeroes to end
```

```python
def moveZeroes(nums):
    insert_pos = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[insert_pos] = nums[i]
            insert_pos += 1
    while insert_pos < len(nums):
        nums[insert_pos] = 0
        insert_pos += 1

arr = [0, 1, 0, 1, 10, 20, 30, 1, 2]
moveZeroes(arr)
print(arr)
```
