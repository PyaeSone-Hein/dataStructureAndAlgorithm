```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, data):
        node = Node(data)
        if self.tail:
            self.tail.next = node
            self.tail = node
        else:
            self.head = node
            self.tail = node

    def append_at_a_location(self, data, index):
        current = self.head
        prev = None
        node = Node(data)
        count = 0

        while current:
            if count == index:  # <-- check against index
                if prev is None:   # inserting at head
                    node.next = current
                    self.head = node
                else:              # inserting in middle
                    node.next = current
                    prev.next = node
                return
            count += 1
            prev = current
            current = current.next

        if count == index:  # inserting at end
            self.tail.next = node
            self.tail = node
        elif count < index:
            print("The list has less number of elements")

# ------------------ TEST -------------------
words = SinglyLinkedList()
words.append('egg')
words.append('ham')
words.append('spam')

print("Before insertion:")
current = words.head
while current:
    print(current.data)
    current = current.next

words.append_at_a_location('new', 2)

print("\nAfter insertion at index 2:")
current = words.head
while current:
    print(current.data)
    current = current.next
```
