```python
def binarySearch(arr, l, r, x):
    while l <= r:
        mid = l + (r - l)//2

        if arr[mid] == x:
            return mid

        elif arr[mid] < x:
            l = mid + 1

        else:
            r = mid -1

    return -1

a = [7,9,11,4,3,2,1,5,6]
arr = sorted(a)
x = 7

result = binarySearch(arr, 0, len(arr)-1, x)
print("Found at index:", result)
```

```python
def power(x, n):
    if n == 0:
        return 1

    half = power(x, n//2)
    if n % 2 == 0:
        return half * half

    else:
        return x * half * half

x = 4
n = 5
result = power(x, n)
print(result)
```

```python
def lucas(n):
    if n == 0:
        return 2
    elif n == 1:
        return 1

    return lucas(n-1) + lucas(n-2)

```
