```python
def findOddOneOut(dataList):

    for i, j in enumerate(dataList):

        if j % 2 != 0:
            print(f'Find odd number at index{i}')
            return j

        else:
            print(f'Checked index {i}: even. Continue')

    return 'Error no odd number was found.'

dl = [2, 4, 3, 8, 10, 5, 14]
odd_number = findOddOneOut(dl)
print("\n--- Result ---")
print(f"Input List: {dl}")
print(f"The odd integer (Output): {odd_number}")
```

```python
def findBook(isbnList, target):
    left = 0
    right = len(isbnList) - 1

    while left <= right:
        mid = (left + right) // 2
        print(f'Checking middle index {mid}: {isbnList[mid]}')

        if isbnList[mid] == target:
            print(f'Book found at index {mid}')
            return mid

        elif isbnList[mid] < target:
            print(f"Target > {isbnList[mid]}. Searching right side.")
            left = mid + 1

        else:
            print(f"Target < {isbnList[mid]}. Searching left side.")
            right = mid - 1

    print("Book not found on the shelf.")
    return -1


bookshelf = [101, 102, 103, 105, 107, 110, 115]
target_isbn = 105

result = findBook(bookshelf, target_isbn)

print("\n--- Result ---")
print(f"Bookshelf: {bookshelf}")
print(f"Target ISBN: {target_isbn}")
print(f"Output: {result}")
```

```python
def find_missing_number(nums):
    unknown_position = -1
    for i in range(len(nums)):
        if nums[i] == -1:
            unknown_position = i
            print(f"Unknown value (-1) found at index {i}")
            break
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2
        if mid == unknown_position:
            if nums[right] == right:
                right = mid - 1
            else:
                left = mid + 1
            continue

        print(f"Checking index {mid}: value = {nums[mid]}")

        if nums[mid] == mid:
            left = mid + 1
        else:
            right = mid - 1
    print(f"Missing number identified: {left}")
    return left

data = [0, 1, 2, -1, 4, 5]
result = find_missing_number(data)

print("\n--- Result ---")
print(f"Input List: {data}")
print(f"Missing Number (Output): {result}")
```

```python
def linearSearch(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i

    return -1

arr = [10, 30, 50, 60, 70, 90]
target = 70

result = linearSearch(arr, target)

if result != -1:
    print(f'Target is found at index {result}')

else:
    print('Target not exist in a list')
```

```python
def binarySearch(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = low + (high-low)//2

        if arr[mid] == target:
            return mid

        elif arr[mid] > target:
            high = mid -1

        else:
            low = mid + 1

    return -1

arr = [11, 23, 34, 45, 56, 78, 89 ]
target = 45
result = binarySearch(arr, target)

if result != -1:
    print(f'Target is found at index {result}')

else:
    print('Target not exist in a list')
```

```python
def findMissingNumber(arr):
    unknown_pos = arr.index(-1)

    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if mid == unknown_pos:
            left = mid + 1
            continue

        expected_value = mid
        if mid > unknown_pos:
            expected_value -= 1

        if arr[mid] == expected_value:
            left = mid + 1
        else:
            right = mid - 1

    return left

arr = [0, 1, 2, -1, 4, 5]
print(findMissingNumber(arr))
```

```python
def findMissingNumber_Correct(arr):
    L = len(arr)
    # 1. Calculate Expected Sum (0 + 1 + ... + L-1)
    expected_sum = L * (L - 1) // 2
    
    # 2. Calculate Actual Sum of elements (excluding the -1)
    actual_sum = 0
    for x in arr:
        # Only add non-negative numbers
        if x != -1:
            actual_sum += x
            
    # The missing number is the difference
    missing_number = expected_sum - actual_sum
    
    return missing_number
arr = [0, 1, 2, -1, 4, 5]
print(findMissingNumber_Correct(arr))
```
