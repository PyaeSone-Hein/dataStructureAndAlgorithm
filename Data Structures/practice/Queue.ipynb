```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class myqueue():
    def __init__(self):
            self.front = None
            self.rear = None

    #add data (rear)
    def enqueue(self, data):
        if self.rear is None:
            self.front = self.rear = Node(data)

        else:
            self.rear.next = Node(data)
            self.rear = self.rear.next

    #remove data(front)
    def dequeue(self):
        if self.front is None:
            return None

        else:
            temp = self.front
            self.front = self.front.next
            if self.front is None: #after removing the queue can become none
                self.rear = None

            return temp.data #remove data

    def isEmpty(self):
        return self.front is None

    def getSize(self):
        temp = self.front #traverse from front
        size = 0
        while temp:
            size += 1
            temp = temp.next #move to next node
        return size

    def display(self):
        if self.front is None:
            return None

        else:
            temp = self.front 
            while temp:
                print(temp.data, end =  '-->')
                temp = temp.next

            print()

#test
queue = myqueue()
queue.enqueue(200)
queue.enqueue(300)
queue.enqueue(600)
queue.display()
print('\n')

print(f'{queue.dequeue()}\n')
queue.display()
```

```python
import time

def simulate_printer_queue(print_jobs):
    printer_queue = myqueue()

    for job in print_jobs:
        pages, delay = job
        time.sleep(delay)
        print(f'Add jon with {pages} pages to the queue')
        printer_queue.enqueue(pages)
        print(f"Number of Jobs in Queue: {printer_queue.getSize()}")
        
    print("----------------------------")

    while not printer_queue.isEmpty():
        current_job = printer_queue.dequeue()
        print(f"Printing job with {current_job} pages...")
        time.sleep(current_job * 0.1)  # Simulate time taken to print the job
        print(f"Job with {current_job} pages completed.")
        print(f"Number of Jobs in Queue: {printer_queue.getSize()}")

# Test the function
# (number of pages, delay in seconds before arrival)
print_jobs = [(5, 1), (3, 2), (8, 3)]  
simulate_printer_queue(print_jobs)
```

```python
class Node():
    def __init__(self, data, priority):
        self.data = data
        self.priority = priority
        self.next = None

class priorityQueue():
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data, priority):
        new_node = Node(data, priority)

        if self.front is None:
            self.front = self.rear = new_node

            return

        if priority > self.front.priority:
            new_node.next = self.front
            self.front = new_node
            return

        current = self.front
        while current.next and current.next.priority >= priority:
            current = current.next

        new_node.next = current.next
        current.next = new_node

        if new_node.next is None:
            self.rear = new_node

    # Dequeue method
    def dequeue(self):
        if self.front is None:
            print("Queue is empty. Nothing to dequeue.")
            return None

        temp = self.front
        self.front = self.front.next

        if self.front is None:
            self.rear = None

        return temp.data

    # Peek method
    def peek(self):
        if self.front is None:
            print("Queue is empty.")
            return None
        return self.front.data

    # Display method
    def display(self):
        if self.front is None:
            print("Queue is empty.")
            return

        current = self.front
        print("Front -> ", end="")
        while current:
            print(f"[{current.data} (P{current.priority})]", end=" -> ")
            current = current.next
        print("Rear")


# ----------------------------
# Test Code
# ----------------------------
pq = priorityQueue()
pq.enqueue(99, 5)
pq.enqueue(23, 3)
pq.enqueue(1, 2)
pq.enqueue(6, 4)

pq.display()

print("\nDequeued:", pq.dequeue())
pq.display()

print("\nPeek:", pq.peek())          
```
