```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None

    def push(self, data):
        if self.head is None:
            self.head = Node(data)

        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node


    def pop(self):
        if self.head is None:
            return None
        else:
            popped = self.head.data
            self.head = self.head.next

            return popped

    def peek(self):
        if self.head is None:
            return None

        else:
            return self.head.data

    def is_empty(self):
        return self.head is None

    def display(self):
        current = self.head
        while current is not None:
            print(current.data, end = ' ->')
            current = current.next
        print()

#test
stack = Stack()

stack.push(200)
stack.push(500)
stack.push(600)

stack.display()

print(f'\n Top element is :{stack.peek()}')
print(stack.pop())
print(f'{stack.display()}')
```

```python
def reverse_string_with_stack(input_string):
    stack = []

    for char in input_string:
        stack.append(char)

    reversed_string = ''
    while stack:
        reversed_string += stack.pop()

    return reversed_string

input_string = input('Enter a sentence: ')
output = reverse_string_with_stack(input_string)
print(output)
```

```python
import queue

def convert_num_to_bin(n):
    stack1 = queue.LifoQueue(maxsize=0)

    while n > 0:
        stack1.put(n % 2)
        n = n // 2

    while not stack1.empty():
        print(stack1.get(), end='')

convert_num_to_bin(121)
print() 
```

```python
def evaluate_postfix(expression):
    stack = Stack()

    tokens = expression.split()

    for token in tokens:
        if token.isdigit():
            stack.push(int(token))

        else:
            b = stack.pop()
            a = stack.pop()

            if token == '+':
                stack.push(a+b)

            elif token == '-':
                stack.push(a-b)

            elif token == '*':
                stack.push(a*b)

            elif token == '/':
                stack.push(a/b)

        return stack.peek()

expression = "3 8 2 / + 5 -"
result = evaluate_postfix(expression)
print(f"The result of the postfix expression '{expression}' is:{result}")
```

```python
def is_balanced(expression):
    stack = []

    opening_brackets = {'(', '[', '{'}

    matching_bracket = {')': '(', ']': '[', '}': '{'}

    for char in expression:

        if char in opening_brackets:
            stack.append(char)

        elif char in matching_bracket:

            if len(stack) == 0:
                return False

            last_opener = stack.pop()

            if last_opener != matching_bracket[char]:
                return False

    return len(stack) == 0

print(f"'()': {is_balanced('()')}")
```
