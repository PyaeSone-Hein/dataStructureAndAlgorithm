```python
from heapq import heappush, heappop, heapify
```

```python
class MinHeap:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i-1)//2

    def insertKey(self, k):
        heappush(self.heap, k)

    def getMin(self):
        return self.heap[0]

    def extractMin(self):
        return heappop(self.heap)

    def descreaseKey(self, i, new_val):
        self.heap[i] = new_val
        while (i > 0 and self.heap[self.parent(i)] > self.heap[i]):
            self.heap[i] , self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
            i = self.parent(i)

h = MinHeap()

# Insert elements
h.insertKey(3)
h.insertKey(2)
h.insertKey(15)
h.insertKey(5)
h.insertKey(4)
h.insertKey(45)

print("Heap array:", h.heap)

# Get minimum
print("Minimum element:", h.getMin())

# Extract minimum
print("Extracted min:", h.extractMin())
print("Heap after extracting min:", h.heap)

# Decrease key at index 2
h.descreaseKey(2, 1)
print("Heap after decreaseKey:", h.heap)

print("New minimum:", h.getMin())
```
